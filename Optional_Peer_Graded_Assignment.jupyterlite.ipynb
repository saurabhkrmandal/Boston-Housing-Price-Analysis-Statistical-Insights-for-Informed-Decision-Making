```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

#### Read the dataset in the csv file from the URL


```python
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
resp = await fetch(URL)
boston_url = io.BytesIO((await resp.arrayBuffer()).to_py())
```

```python
boston_df=pd.read_csv(boston_url)
```

#### Add your code below following the instructions given in the course to complete the peer graded assignment


```python
# Task 1: Become Familiar with the Dataset

# 1. Display the first few rows of the dataset to become familiar with it
print("First few rows of the dataset:")
display(boston_df.head())  # Using display() ensures output visibility in Jupyter

# 2. Get summary statistics for the dataset
print("\nDataset summary (Descriptive statistics):")
display(boston_df.describe())  # Describes key statistics for numerical columns

# 3. Check for missing values
print("\nCheck for missing values:")
display(boston_df.isnull().sum())  # Shows the count of missing values in each column


# Task 2a: Boxplot for the "Median value of owner-occupied homes" (MEDV)
plt.figure(figsize=(10, 6))
sns.boxplot(y='MEDV', data=boston_df)
plt.title('Boxplot for Median Value of Owner-Occupied Homes (MEDV)')
plt.ylabel('MEDV ($1000)')
plt.show()

# Explanation for the boxplot:
# The boxplot shows the distribution of the median value of owner-occupied homes (MEDV).
# It highlights the central tendency and spread, with potential outliers shown as points outside the whiskers.

# Task 2b: Bar plot for the Charles River variable (CHAS)
plt.figure(figsize=(10, 6))
sns.countplot(x='CHAS', data=boston_df)
plt.title('Bar Plot for Charles River Variable (CHAS)')
plt.xlabel('Bounded by Charles River (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()

# Explanation for the bar plot:
# This bar plot shows the number of homes that are bounded by the Charles River (CHAS = 1) versus those that are not (CHAS = 0).

# Task 2c: Boxplot for MEDV vs AGE group (discretized AGE)
# Discretize the AGE variable into three groups
boston_df['AGE_group'] = pd.cut(boston_df['AGE'], bins=[0, 35, 70, 100], labels=['<=35', '35-70', '>70'])

plt.figure(figsize=(10, 6))
sns.boxplot(x='AGE_group', y='MEDV', data=boston_df)
plt.title('Boxplot for Median Value of Homes (MEDV) by Age Group')
plt.xlabel('Age Group')
plt.ylabel('MEDV ($1000)')
plt.show()

# Explanation for the boxplot:
# This boxplot compares the median value of homes (MEDV) across different age groups.
# The age groups are: homes built before 35 years, between 35 to 70 years, and more than 70 years ago.

# Task 2d: Scatter plot of Nitric oxide concentrations (NOX) vs non-retail business acres (INDUS)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='NOX', y='INDUS', data=boston_df)
plt.title('Scatter Plot: Nitric Oxide Concentration (NOX) vs Non-retail Business Acres (INDUS)')
plt.xlabel('NOX (parts per 10 million)')
plt.ylabel('INDUS (Proportion of non-retail business acres)')
plt.show()

# Explanation for the scatter plot:
# The scatter plot visualizes the relationship between Nitric oxide concentrations (NOX) and non-retail business acres (INDUS).
# A positive correlation suggests that as NOX levels increase, the proportion of non-retail business acres also increases.

# Task 2e: Histogram for the Pupil-to-Teacher Ratio (PTRATIO)
plt.figure(figsize=(10, 6))
sns.histplot(boston_df['PTRATIO'], bins=10, kde=False)
plt.title('Histogram for Pupil-to-Teacher Ratio (PTRATIO)')
plt.xlabel('PTRATIO')
plt.ylabel('Frequency')
plt.show()

# Explanation for the histogram:
# This histogram displays the distribution of the pupil-to-teacher ratio across different towns.
# It helps to visualize the frequency of different PTRATIO values in the dataset.

# Task 3a: T-test for independent samples (MEDV vs CHAS)
from scipy import stats

# Hypothesis:
# H0: There is no significant difference in the median value of houses between those bounded by the Charles River and those that are not.
# H1: There is a significant difference in the median value of houses.

chas_1 = boston_df[boston_df['CHAS'] == 1]['MEDV']
chas_0 = boston_df[boston_df['CHAS'] == 0]['MEDV']

t_stat, p_value = stats.ttest_ind(chas_1, chas_0, equal_var=False)
print(f"T-test result: t_stat = {t_stat}, p_value = {p_value}")

# Conclusion:
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in the median value of houses bounded by the Charles River.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the median value of houses.")

# Task 3b: ANOVA for MEDV vs AGE group
# Hypothesis:
# H0: There is no significant difference in the median values of houses across different age groups.
# H1: There is a significant difference in the median values of houses across age groups.

anova_result = ols('MEDV ~ AGE_group', data=boston_df).fit()
anova_table = sm.stats.anova_lm(anova_result, typ=2)
print("\nANOVA test result for MEDV vs AGE group:")
print(anova_table)

# Conclusion:
if anova_table['PR(>F)'][0] < 0.05:
    print("Reject the null hypothesis: There is a significant difference in the median value of houses across age groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the median value of houses across age groups.")

# Task 3c: Pearson Correlation between NOX and INDUS
# Hypothesis:
# H0: There is no significant correlation between NOX and INDUS.
# H1: There is a significant correlation between NOX and INDUS.

corr_coef, p_value_corr = stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
print(f"\nPearson Correlation result: corr_coef = {corr_coef}, p_value = {p_value_corr}")

# Conclusion:
if p_value_corr < 0.05:
    print("Reject the null hypothesis: There is a significant correlation between NOX and INDUS.")
else:
    print("Fail to reject the null hypothesis: There is no significant correlation between NOX and INDUS.")

# Task 3d: Regression analysis for DIS vs MEDV
# Hypothesis:
# H0: There is no significant impact of an additional weighted distance on the median value of homes.
# H1: There is a significant impact of an additional weighted distance on the median value of homes.

regression_result = ols('MEDV ~ DIS', data=boston_df).fit()
print("\nRegression analysis result for MEDV vs DIS:")
print(regression_result.summary())

# Conclusion:
if regression_result.pvalues['DIS'] < 0.05:
    print("Reject the null hypothesis: The weighted distance to Boston employment centers has a significant impact on the median value of homes.")
else:
    print("Fail to reject the null hypothesis: The weighted distance to Boston employment centers does not have a significant impact on the median value of homes.")
```
